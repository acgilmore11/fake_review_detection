```python
import pandas as pd
import spacy
import numpy as np
import re
from math import sqrt, pow, exp
from statistics import mean
import datetime as dt
df = pd.read_csv("metadata.csv")
df.columns=['user_id','prod_id','rating','label','date']
df['Date_pd_st'] = pd.to_datetime(df['date'])
df['Date_pd_en']=df['Date_pd_st'].shift(-1).dropna()
df['Date_Entropy']=abs(df['Date_pd_en']-df['Date_pd_st'])/np.timedelta64(1,'D')
df['Date_Entropy']=df['Date_Entropy'].fillna(0).astype(int)
df=df[['user_id','Date_Entropy']]
df
```
