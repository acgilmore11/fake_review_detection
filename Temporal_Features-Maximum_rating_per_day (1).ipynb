```python
import pandas as pd
import numpy as np
import re
from math import sqrt, pow, exp
from statistics import mean
import datetime as dt

cols_meta = ["user_id", "prod_id", "rating", "label", "date"]
df = pd.read_csv("metadata.csv", names = cols_meta)
#print(df)

df['Date_pd_st'] = pd.to_datetime(df['date'])
#df=df[['user_id','Date_pd_st']]
df['Date_pd_en']=df['Date_pd_st'].shift(-1)
df['Date_Entropy']=abs(df['Date_pd_en']-df['Date_pd_st'])/np.timedelta64(1,'D')
df['Date_Entropy']=df['Date_Entropy'].fillna(0).astype(int)
#df=df[['user_id','Date_Entropy']]
#df
#table = pd.merge(table,df, left_on=['user_id'],right_on=['user_id'], validate = '1:m')
#table
table = pd.concat([meta_data, df["Date_Entropy"]], axis = 1).dropna()
table
```

```python

import pandas as pd
import numpy as np
import re
from math import sqrt, pow, exp
from statistics import mean
import datetime as dt
cols_meta = ["user_id", "prod_id", "rating", "label", "date"]
df = pd.read_csv("metadata.csv", names = cols_meta)
#print(df)
df=df[['user_id','date']]
df['Date_int'] = pd.to_datetime(df['date']).astype(np.int64)/1000000000  #nanoseconds to seconds
df['Date_var']=(((df['Date_int'] - df['Date_int'].mean()).abs())**2)
table = pd.concat([meta_data, df["Date_var"]], axis = 1).dropna()
table
```
